```python
import math
import numpy as np
import matplotlib.pyplot as plt
def h(x):
    res=(1/np.sqrt(2*np.pi))*np.exp((-1/2)*x**2)
    return res

# scalar input
print(h(3))  

# vector input
x=np.linspace(-5,5,51)
# print(x)
y=h(x)
# print(y)
plt.plot(x,y)

```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
def s(x,n):
    r=0
    for j in range(n+1):
        r += ((((-1)**j)*(x**(2*j+1)))/(math.factorial(2*j+1)))
    
    return r

x=np.linspace(0,4*(np.pi),35)
y=[s(i,35) for i in x]
# print(x)
# print(y)


plt.plot(x,y)
```
